```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
vendas = pd.read_excel("/content/varejo.xlsx")
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP','Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace("_","")
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'].fillna('MS',inplace=True)
vendas.info()
```

```python
media_preco = vendas['Preço'].mean()

vendas['Preço'].fillna(media_preco, inplace=True)
vendas.info()
```

```python
vendas[['Preço','Preço_com_frete']].describe()
```

```python
preco_errado = vendas.query('Preço > Preço_com_frete')
venda_correto = vendas.query('Preço < Preço_com_frete')
```

```python
venda_correto.describe()
```

```python
venda_correto.head()
```

```python
venda_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'" ).describe()
```

```python
venda_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False)
```

```python
round(venda_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
venda_correto.groupby('mes').idcompra.nunique().sort_values(ascending=False)
```

```python
venda_correto['mes'] = venda_correto['Data'].dt.month
venda_correto.head()
```

```python
cliente = pd.read_excel('/content/cliente_varejo.xlsx')
```

```python
cliente.info()
```

```python
cliente = cliente.astype({'renda':'float'})
```

```python
vendas_cliente = venda_correto.merge(cliente, how='left', on='cliente_Log')
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idcanal_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_bandeira['bandeira'],agg_idcanal_bandeira['idade'],color="green")

plt.ylabel('Média de idade')

plt.title('Idade média por bandeira', loc="center",fontsize=18,fontweight="bold",color='black');
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda['idcanalvenda'],agg_idcanal_renda['renda'],color="green")

plt.ylabel('Renda média')

plt.title('Renda média por canal de venda', loc="center",fontsize=18,fontweight="bold",color='black');
```

```python
venda_por_data = venda_correto.groupby("Data").idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(7,4))

plt.xlabel("Data da Venda")

plt.ylabel("Quantidade de Vendas")

x = venda_por_data["Data"]

y = venda_por_data["idcompra"]

plt.plot(x,y, color="green",linewidth = "2")

plt.show()
```

```python
agg_dept_preco = round(venda_correto.groupby("Nome_Departamento")['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)

agg_dept_preco
```

```python
fig = plt.figure(figsize=(35,20))

plt.bar(agg_dept_preco['Nome_Departamento'],agg_dept_preco['Preço_com_frete'],color="green")

plt.ylabel('Preço média')

plt.title('Preço médio por departamento', loc="center",fontsize=18,fontweight="bold",color='black');
```
